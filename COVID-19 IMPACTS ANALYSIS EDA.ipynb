```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
Covid_19=pd.read_csv("Desktop\\transformed_data.csv")
```

```python
Covid_19=Covid_19.dropna()
```

```python
Covid_19.shape
```

```python
Covid_19.groupby("COUNTRY")["TC","TD","STI","HDI"].agg(np.mean).plot(kind='line',rot=31)
```

```python
Covid_19.groupby("COUNTRY")["TD"].agg(np.mean).plot(kind='line',rot=31)
```

```python
Covid_19.pivot_table(values="TC",index="COUNTRY",columns="DATE")
```

```python
country=Covid_19.groupby("COUNTRY")["TD","TC"].max()
TOP_10_Country=country.iloc[0:11,:]
country_list = TOP_10_Country.index.tolist()
plt.title("TOP 10 Country")
ax=sns.barplot(data=Covid_19[Covid_19['COUNTRY'].isin(country_list)],x="COUNTRY",y="TD",linewidth=2,edgecolor='black')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

```python
country=Covid_19.groupby("COUNTRY")["TD","TC"].max()
TOP_10_Country=country.iloc[0:11,:]
TOP_10_Country.plot(kind="bar",title="TOP 10 Country")
```

```python
sns.lineplot(data=Covid_19[Covid_19['COUNTRY'].isin(['Afghanistan','Albania','Algeria'])],x="DATE",y="TC",hue="COUNTRY")
```
